```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("nba.csv")
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python

df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df['College'].fillna(df['College'].mode()[0], inplace=True)
df['Height'].fillna(df['Height'].mode()[0], inplace=True)
df['Name'].fillna(df['Name'].mode()[0], inplace=True)
df['Team'].fillna(df['Team'].mode()[0], inplace=True)
df['Number'].fillna(df['Number'].mean(), inplace=True)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Weight'].fillna(df['Weight'].mean(), inplace=True)
```

```python
#min-max scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
cols_to_scale = ['Salary', 'Age', 'Weight', 'Number']
```

```python
scaler = MinMaxScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python

sns.histplot(df['Salary'], bins=30, kde=True, color='green')
plt.title("Salary Distribution of NBA Players")
plt.xlabel("Normalized Salary (0 to 1)")
plt.ylabel("Number of Players")
plt.show()
```

```python
df.avg('Salary')
```

```python
sns.barplot(x='Position', y='Age', data=df)
plt.title("Average Age by Position")
plt.ylabel("Normalized Age (0 to 1)")
plt.show()
```

```python
sns.barplot(x='Name', y='Age', data=df)
plt.title("Average Age by Position")
plt.ylabel("Normalized Age (0 to 1)")
plt.show()
```

```python
df['Team'].value_counts().plot(kind='bar', color='purple')
plt.title("Number of Players per Team")
plt.xlabel("Team")
plt.ylabel("Player Count")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.heatmap(df[cols_to_scale].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(df)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python

salary_by_team = df.groupby('Team')['Salary'].sum().sort_values(ascending=False).head(6)

plt.pie(salary_by_team, autopct='%1.1f%%', labels=salary_by_team.index, startangle=140)
plt.title('Total Salary Distribution by Top 6 Teams')
plt.axis('equal')  
plt.show()
```
